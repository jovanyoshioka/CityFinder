### Distinguish Duplicate Weather Column Names
`df_tavg`, `df_prcp`, and `df_snow` have the same column names (Jan, Feb, etc.). When merging, won't be able to distinguish which one is temperature, precipitation, and snow.

```python
def add_prefix_except_fips(df, prefix, blacklist):
    df.columns = ['{}{}'.format('' if col in blacklist else prefix, col) for col in df.columns]
    return df
```

### Cleaning FIPS

```python
def add_char(string, char, position): 
    # Convert the string to a list of characters 
    string_list = list(string) 
     
    # Insert the character at the specified position 
    string_list.insert(position, str(char)) 
    
    # Join the list back into a string 
    new_string = "".join(string_list)
     
    return new_string

def clean_statefips(statefips):
    for i in range(len(statefips)): 
        if len(str(statefips[i])) != 2:
            statefips[i] = add_char(str(statefips[i]),0,0)
        else: 
            statefips[i] = str(statefips[i])
    return statefips
            
def clean_countyfips(countyfips):
    for i in range(len(countyfips)):
        if len(str(countyfips[i])) == 1:
            countyfips[i] = add_char(str(countyfips[i]),0,0)
            countyfips[i] = add_char(str(countyfips[i]),0,0)
        elif len(str(countyfips[i])) == 2: 
            countyfips[i] = add_char(str(countyfips[i]),0,0)
        else:
            countyfips[i] = str(countyfips[i])
    return countyfips
```

### Standardize Identifying Column Names

```python
state_fips_codes = {}
county_fips_codes = {}
def create_county_state_dicts(fips):
    county_state = df_fips.loc[df_fips['fips'] == fips].iloc[0]
    county = county_state['name']
    state = county_state['state']
    
    state_fips = fips[:2]
    
    county_fips_codes[fips] = county
    state_fips_codes[state_fips] = state
    
    return fips
    
    
# Note: converters={'FIPS': str} to prevent leading zeros from being removed.
df_fips = pd.read_csv('data/state_and_county_fips.csv', converters={'fips': str})
df_fips['fips'] = df_fips['fips'].apply(lambda fips: fips.zfill(5))
df_fips['fips'] = df_fips['fips'].apply(create_county_state_dicts)
```

```python
def convert_state_fips(fips):
    # Note: May be `null` if fips code not included in `state_and_county_fips.csv`.
    if fips in state_fips_codes:
        return state_fips_codes[fips]
    else:
        return 'null'

def convert_county_fips(fips):
    # Note: May be `null` if fips code not included in `state_and_county_fips.csv`.
    if fips in county_fips_codes:
        return county_fips_codes[fips]
    else:
        return 'null'
```
