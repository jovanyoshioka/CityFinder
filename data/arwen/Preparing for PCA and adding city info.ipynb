```python
def add_char(string, char, position): 
    # Convert the string to a list of characters 
    string_list = list(string) 
     
    # Insert the character at the specified position 
    string_list.insert(position, str(char)) 
    
    # Join the list back into a string 
    new_string = "".join(string_list)
     
    return new_string 

for i in range(len(df1.FIPS)): 
    if len(df1.FIPS[i]) != 5:
        
        df1.FIPS[i] = add_char(df1.FIPS[i],0,0)
        
    else: 
        df1.FIPS[i] = df1.FIPS[i]

df1
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

import sklearn

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("df_for_PCA.csv")
```

```python
df
```

```python
df = df.drop(columns={"Unnamed: 0"})
df
```

```python
df = df.fillna(0)
```

```python
pca = PCA(n_components=1)
pca.fit_transform(df)
```

```python
std_scaler = StandardScaler()
scaled_df = std_scaler.fit_transform(df)
```

```python
pca = PCA(n_components=1)
pca.fit_transform(scaled_df)
```

```python
pd.DataFrame(pca.components_,columns=df.columns,index = ['PC-1'])
```

```python
pca = PCA(n_components=1)
pca.fit_transform(df)
```

```python
PCA_done = pd.DataFrame(pca.components_,columns=df.columns,index = ['PC-1'])
```
