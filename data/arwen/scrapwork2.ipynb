```python
import pandas as pd
import numpy as np
```

```python
def add_char(string, char, position): 
    # Convert the string to a list of characters 
    string_list = list(string) 
     
    # Insert the character at the specified position 
    string_list.insert(position, str(char)) 
    
    
    #string_list = str(string_list)
    # Join the list back into a string 
    new_string = "".join(string_list)
     
    return new_string 
```

```python
#Taxes

taxes = pd.read_csv("21incyallnoagi.csv", encoding='ISO-8859-1' )

taxes.STATEFIPS = taxes.STATEFIPS.astype('string')
taxes.COUNTYFIPS = taxes.COUNTYFIPS.astype('string')
 

for i in range(len(taxes.STATEFIPS)): 
    if len(taxes.STATEFIPS[i]) != 2:
        
        taxes.STATEFIPS[i] = add_char(taxes.STATEFIPS[i],0,0)
        
    else: 
        taxes.STATEFIPS[i] = taxes.STATEFIPS[i]
        
        
        
for i in range(len(taxes.COUNTYFIPS)):
    if len(taxes.COUNTYFIPS[i]) == 1:
        taxes.COUNTYFIPS[i] = add_char(taxes.COUNTYFIPS[i],0,0)
        taxes.COUNTYFIPS[i] = add_char(taxes.COUNTYFIPS[i],0,0)
    elif len(taxes.COUNTYFIPS[i]) == 2: 
        taxes.COUNTYFIPS[i] = add_char(taxes.COUNTYFIPS[i],0,0)
    else:
        taxes.COUNTYFIPS[i] = taxes.COUNTYFIPS[i]
        
taxes["FIPS"] = taxes["STATEFIPS"] + taxes["COUNTYFIPS"]

taxes = taxes[["STATE", "COUNTYNAME",
            "N1", "A00100", "A02650", "A18425",
             "A18450", "A18500", "A18800", "A18300", "A10600", "FIPS"]]
taxes = taxes.rename(columns={"COUNTYNAME":"COUNTY","N1":"Num_Returns",
                             "A00100":"adj_gross_income", "A02650":"total_income",
                              "A18425": "st_local_income_tax","A18450":"st_local_gen_sales_tax", 
                              "A18500":"realestate_tax", "A18800":"personal_prop_tax", 
                              "A18300":"total_taxes_paid", "A10600":"total_tax_payments_amt"})
taxes["YEAR"] = 2021
```
